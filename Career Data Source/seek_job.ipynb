```python
# find job id(s)

import json

# jobs 1
data1 = json.load(open('jobs1.json'))
jobs1 = data1['data']['docs']

job_ids1 = [x['id'] for x in jobs1]

# jobs 2
data2 = json.load(open('jobs2.json'))
jobs2 = data2['data']['docs']

job_ids2 = [x['id'] for x in jobs2]

# jobs 3
data3 = json.load(open('jobs3.json'))
jobs3 = data3['data']['docs']

job_ids3 = [x['id'] for x in jobs3]

final_id = job_ids1 + job_ids2 + job_ids3
len(final_id)
```

```python
# retrieve job data
import requests
import pandas as pd
import json

dfs = []

# URL of the JSON data
for job_id in final_id :
    url = f'https://api.sejutacita.id/v1/job-portal/job/{job_id}'  # Replace this with the actual URL

    # Make the request to get the JSON data
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        data = response.json()
        
        # Extract the relevant data from the JSON
        job_data = data['data']['result']
        
        # Convert the JSON data to a pandas DataFrame
        df = pd.json_normalize(job_data)

        # Append the DataFrame to the list
        dfs.append(df)

final_df = pd.concat(dfs, ignore_index=True)
final_df
```

```python
final_df.to_csv('job_data.csv')
```
